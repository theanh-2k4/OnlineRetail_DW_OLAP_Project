```python
import pandas as pd

df = pd.read_csv('online_retail_2.csv')

# Trích xuất danh sách các quốc gia duy nhất
unique_countries = df['Country'].unique()

# Sắp xếp theo bảng chữ cái để dễ quan sát
unique_countries.sort()

print("Danh sách các quốc gia xuất hiện trong dữ liệu:")
print(unique_countries)
```

```python
# Định nghĩa từ điển Mapping dựa trên danh sách quốc gia thực tế của bạn
region_mapping = {
    # Nhóm 1: UK & Ireland (Thị trường trọng điểm)
    'United Kingdom': 'UK & Ireland',
    'EIRE': 'UK & Ireland',
    'Channel Islands': 'UK & Ireland',
    
    # Nhóm 2: Western Europe (Châu Âu lục địa)
    'France': 'Western Europe',
    'Germany': 'Western Europe',
    'Netherlands': 'Western Europe',
    'Belgium': 'Western Europe',
    'Switzerland': 'Western Europe',
    'Austria': 'Western Europe',
    'European Community': 'Western Europe',
    
    # Nhóm 3: Northern Europe (Bắc Âu)
    'Sweden': 'Northern Europe',
    'Norway': 'Northern Europe',
    'Denmark': 'Northern Europe',
    'Finland': 'Northern Europe',
    'Iceland': 'Northern Europe',
    
    # Nhóm 4: Southern Europe (Nam Âu)
    'Spain': 'Southern Europe',
    'Italy': 'Southern Europe',
    'Portugal': 'Southern Europe',
    'Greece': 'Southern Europe',
    'Malta': 'Southern Europe',
    'Cyprus': 'Southern Europe',
    
    # Nhóm 5: Central & Eastern Europe (Trung và Đông Âu)
    'Poland': 'Eastern Europe',
    'Lithuania': 'Eastern Europe',
    'Czech Republic': 'Eastern Europe',
    
    # Nhóm 6: APAC (Châu Á - Thái Bình Dương)
    'Australia': 'APAC',
    'Japan': 'APAC',
    'Singapore': 'APAC',
    'Thailand': 'APAC',
    'Hong Kong': 'APAC',
    'Korea': 'APAC',
    
    # Nhóm 7: Middle East & Africa (Trung Đông & Phi)
    'United Arab Emirates': 'Middle East',
    'Israel': 'Middle East',
    'Saudi Arabia': 'Middle East',
    'Lebanon': 'Middle East',
    'Bahrain': 'Middle East',
    'Nigeria': 'Africa',
    'RSA': 'Africa', # Republic of South Africa
    
    # Nhóm 8: Americas (Châu Mỹ)
    'USA': 'Americas',
    'Canada': 'Americas',
    'Brazil': 'Americas',
    'West Indies': 'Americas',
    'Bermuda': 'Americas'
}

# Áp dụng Mapping vào DataFrame
df['Region'] = df['Country'].map(region_mapping).fillna('Other/Unspecified')

# Kiểm tra kết quả
print(df[['Country', 'Region']].drop_duplicates())
```

```python
# 1. Định nghĩa Mapping từ Region sang Continent
continent_mapping = {
    'UK & Ireland': 'Europe',
    'Western Europe': 'Europe',
    'Northern Europe': 'Europe',
    'Southern Europe': 'Europe',
    'Eastern Europe': 'Europe',
    'APAC': 'Asia & Oceania',
    'Middle East': 'Asia & Africa',
    'Africa': 'Africa',
    'Americas': 'Americas',
    'Other/Unspecified': 'Other'
}

# 2. Áp dụng Mapping vào DataFrame (Dựa trên cột Region vừa tạo)
df['Continent'] = df['Region'].map(continent_mapping).fillna('Other')

# 3. Lấy danh sách Country - Region - Continent duy nhất
country_region_continent = df[['Country', 'Region', 'Continent']].drop_duplicates()

# 4. Lưu thành file Excel với đầy đủ thông tin
country_region_continent.to_excel('country_mapping_full.xlsx', index=False)

# Kiểm tra kết quả
print(country_region_continent.head(10))
print("\nĐã lưu file country_mapping_full.xlsx")
```
